```python
import matplotlib.pyplot as plt
```

```python
# cv 와 matplot 에서의 RGB 레이어가 다른지 확인해보자 
imgBGR = cv2.imread('./puppy.bmp', 1)

if imgBGR is None :
    print('image read failed')
    sys.exit()

imgBGR = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB) # matplot에서는 CV2와 RGB 순서가 반대
plt.imshow(imgBGR)
plt.axis('off')
plt.show()
# cv2.namedWindow('puppy', cv2.WINDOW_AUTOSIZE)
# cv2.imshow('puppy', img)

# while True :
#     if cv2.waitKey() == 27 :
#         break
        
# cv2.destroyAllWindows()
```

```python
imgGray = cv2.imread('./puppy.bmp', -1) # flag 1 = 컬러, 0 = 반전, -1은 흑백

if imgGray is None : 
    print('imgGray read failed')
    sys.exit()
    
plt.subplot(121), plt.imshow(imgBGR), plt.axis('off')
plt.subplot(122), plt.imshow(imgGray), plt.axis('off')
```
