```python
import pandas as pd

exam_data = {"이름" : ['서준', '우현', '인아'],
            '수학' : [90, 80, 70],
            '영어' : [98, 89, 95],
            '음악' : [85, 95, 100],1
            '체육' : [100, 90, 90]}
df = pd.DataFrame(exam_data)
print(df)
print("---------------------------------")
# '이름'열을 새로운 인덱스로 지정하고, df 객체에 변경 사항 반영
df.set_index('이름', inplace=True)
print(df)
print('----------------------------------')
print(df['수학'])
```

```python
import pandas as pd

# 딕셔서리를 정의
dict_data = {'c0':[1,2,3], 'c1':[4,5,6], 'c2':[7,8,9], 'c3':[10,11,12], 'c4':[13,14,15]}

# 딕셔서리를 데이터프레임으로 변환. 인덱스를 [r0, r1, r2]로 지정
df = pd.DataFrame(dict_data, index=['r0', 'r1', 'r2'])
print(df)
print('\n')

# sort : 인덱스로 정렬하는 경우, 특정 칼럼의 값으로 정렬하는 경우 
print(df.sort_index(ascending=False)) # index로 정렬 ascending=False 역으로 정렬
df.sort_values(by='c0', ascending=False) # False = 역? True = 정? 
# .sort_index => 행으로 정렬 .sort_values => 열로 정렬
df.sort_values(by=['c0','c1'], ascending=False)
```

```python
# 인덱스를 [r0, r1, r2, r3, r4]로 재지정
new_index = ['r0', 'r1', 'r2', 'r3', 'r4'] # 추가되는 데이터는 NaN로 입력됨
ndf = df.reindex(new_index)
# ndf = df.reindex(new_index, fill_value=0) # 추가되는 데이터에 초기값 설정
print(ndf)
print('\n')
```

```python

# 판다스의 연산
# 대응되는 값이 없다면 NaN 값 출력
import pandas as pd

# 딕셔너리 데이터로 판다스 시리즈 만들기 
student1 = pd.Series({'국어':100, '영어':80, '수학':90})
student2 = pd.Series({'국어':90, '영어':85, '수학':95, '과학':100})
print(student1)
print('\n')

print(student1/200)
print('\n')
print(student1 + student2) #Series 연산자(+, -, *, /) series 또는 숫자 
print('\n')
print(student1 - student2) 
# 내부적으로 index로 정렬한 후 각각의 index로 계산함 

print(student1.add(student2, fill_value=0)) # add, sub, mul, div(자료가 없으면 원하는 값으로 대체)
# 연산식에 NaN이 포함되면 결과는 NaN으로 나타남 -> 옵션으로 fill_value=원하는 값으로 치환
```

```python
# dataframe 연산, 기존의 모듈에서 dataset을 불러들임 
import seaborn as sns

titanic = sns.load_dataset('titanic')
type(titanic)
df = titanic.loc[ : ,['age','fare']] # 'age'와'fare' 칼럼의 정보만 가져옴 
# 100~200 인덱스 정보만 가져옴
df = df.iloc[100:201]
print(df.head(5))

print(df + 10)
df.add(10, fill_value=0)

df1 = df.head(10)
df2 = df.tail(10)
print(df1.add(df2,fill_value=0))
```

```python
avg_year = int(df.age.sum() / df.age.count())
# 평균나이보다 적은 연령의 자료만 출력 
a = 0
for i,year in enumerate(df.loc[ : ,'age']) : 
    if year < avg_year : 
        a = i+1
        print(year)

print("-------------------")
print(a)
```

```python
# ['age','fare']
avg_age = int(df.age.sum() / df.age.count())
for idx, age in enumerate(df.age) :
    if age < avg_age :
        print("{} : {}".format(df.iloc[idx,0],df.iloc[idx,1]))
```
