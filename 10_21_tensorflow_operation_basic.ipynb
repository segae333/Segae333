### Perceptron 작성

```python

import tensorflow as tf

## data 선언
x_data =[[1.],[2.],[3.],[4.]]
y_data =[[1.],[3.],[5.],[7.]]

print(x_data)
print(y_data)
```

```python
## 평균 0, 분산 1의 파라미터의 정규분포로 부터 값을 가져옴.
# 학습을 통해 업데이트가 되어 변화되는 모델의 파라미터인 w,b를 의미한다.
W= tf.Variable(tf.random.normal((1,1), mean = 0, stddev=1.0))  
b= tf.Variable(tf.random.normal((1,1), mean = 0, stddev=1.0))
print("W : {}".format(W))
print("b : {}".format(b))
```

```python
for j in range(len(x_data)):
    ## data * weight 작성
    WX = tf.matmul([x_data[j]], W)

    ## bias add 작성
    y_hat = tf.add(WX , b)

    ## W와 b로 예측 하기
    print("y_data: , ",y_data[j], "prediction : ", y_hat)
```

# Tensorflow Linear Regression

```python
import tensorflow as tf

## data 선언
x_data =[[1.],[2.],[3.],[4.]]
y_data =[[1.],[3.],[5.],[7.]]

```

```python
## 평균 0, 분산 1의 파라미터의 정규분포로 부터 값을 가져옴.
# 학습을 통해 업데이트가 되어 변화되는 모델의 파라미터인 w,b를 의미한다.
W=tf.Variable(tf.random.normal((1,1),mean=0, stddev=1.0))
b=tf.Variable(tf.random.normal((1,1),mean=0, stddev=1.0))
lr=tf.constant(0.0001) # 보폭
history = np.zeros([2000, 4], 'float32')
```

```python

for i in range(2000):  ## 에폭(전체 데이터를 몇 번 볼거냐)
    total_error = 0

    for j in range(len(x_data)): ## 배치 1(전체 데이터를 몇 개 씩 나눠 볼거냐)
        ## data * weight
        WX =tf.matmul([x_data[j]], W) ## [1.] * [1X1 : w:init]

        ## bias add
        y_hat = tf.add(WX, b) ## y= wx + b

        ## 정답인 Y와 출력값의 error 계산
        error = tf.subtract(y_data[j], y_hat) # true - prediction

        ## 경사하강법으로 W와 b 업데이트.
        ## 도함수 구하기 
        diff_W = tf.matmul(error, [x_data[j]]) # error * x의 합
        diff_b = error

        ##  업데이트할 만큼 러닝레이트 곱
        diff_W = tf.multiply(lr, diff_W)
        diff_b = tf.multiply(lr, diff_b) # lr * error

        ## w, b 업데이트
        W = tf.add(W, diff_W) # W + delta w + lr * x * (error)
        b = tf.add(b, diff_b) # b + delta b + lr * (error)
        #######

        ## 토탈 에러.
        visual_error = tf.square(error)
        total_error = total_error + visual_error

    ## 모든 데이터에 따른 error 값
    print("epoch: ", i, "error : ", total_error/len(x_data))
    history[i,:] = [(total_error/len(x_data))[0], W[0], b[0]]
```

```python
## 학습끝난 후 W와 b로 예측 하기
print(history)
print("W : ", W)
print("b : ", b)
print("input 3 : ", tf.add(tf.multiply(3, W), b))
print("input 5 : ", tf.add(tf.multiply(5, W), b))
```

# Tensorflow Keras

```python
import tensorflow as tf

x_data =[[1.],[2.],[3.],[4.]]
y_data =[[1.],[3.],[5.],[7.]]

test_data = [[9.]]
```

```python
model = tf.keras.Sequential()

model.add(tf.keras.layers.Dense(1, input_dim=1))
model.summary()
```

```python
optimizer = tf.keras.optimizers.SGD(lr=0.01) # 경사하강법 global min 찾아가는 최적의 방법
loss = tf.keras.losses.mse # 예측한 값과 정답간 오차를 계산. mse = mean square error (예측값 - 정답)^2
metrics = tf.keras.metrics.mae # mae = mean absolute error |예측값 - 정답| 
```

```python
# 모델 컴파일
model.compile(loss='mse', optimizer=optimizer, metrics=[metrics])

model.fit(x_data, y_data, epochs=2000, batch_size=1)
```

```python
# 결과를 출력
print(model.weights)
print("test data [9,] 예측값 :", model.predict(test_data))
```
